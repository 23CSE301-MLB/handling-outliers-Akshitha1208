```python
z_scores = np.abs(stats.zscore(df['platelets']))
outliers_z = df[z_scores > 3]
print(outliers_z.shape)
```

```python
Q1 = df['platelets'].quantile(0.25)
Q3 = df['platelets'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers_iqr = df[(df['platelets'] < lower_bound) | (df['platelets'] > upper_bound)]
print(outliers_iqr.shape)
```

```python
df_removed = df[(df['platelets'] >= lower_bound) & (df['platelets'] <= upper_bound)]
print(df_removed.shape)
```

```python
outliers_iqr1 = df_removed[(df_removed['platelets'] < lower_bound) | (df_removed['platelets'] > upper_bound)]
print(outliers_iqr1.shape)
```

```python
df_capped = df.copy()
df_capped['platelets'] = np.where(df_capped['platelets'] > upper_bound, upper_bound,
                                    np.where(df_capped['platelets'] < lower_bound, lower_bound, df_capped['platelets']))
```

```python
df_capped
```

```python
outliers_iqr2 = df_capped[(df_capped['platelets'] < lower_bound) | (df_capped['platelets'] > upper_bound)]
print(outliers_iqr2.shape)
```

```python
df_log = df.copy()
df_log['platelets'] = np.log(df_log['platelets'] + 1)
```

```python
df_log
```

```python
df_imp = df.copy()
median_value = df['platelets'].median()
df_imp.loc[(df_imp['platelets'] > upper_bound) | 
(df_imp['platelets'] < lower_bound), 
'platelets'] = median_value
```

```python
df_imp
```

```python
outliers_iqr4 = df_imp[(df_capped['platelets'] < lower_bound) | (df_imp['platelets'] > upper_bound)]
print(outliers_iqr4.shape)
```
